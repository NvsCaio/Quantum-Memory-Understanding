# Método convencional

## Inicio da implementação 

```python
qubits = 8  # the number of qubits on which the unitaries act
n_shots = 100  # the number of times we can use each unitary
circuits=30
```

## Função para gerar circuitos

```python
np.random.seed(234087)

layers, gates = 10, 10  # the number of layers and gates used in RandomLayers


def generate_circuit(shots):
    """
    generate a random circuit that returns a number of measuement samples
    given by shots
    """
    dev = qml.device("lightning.qubit", wires=qubits, shots=shots)

    @qml.qnode(dev)
    def circuit(ts=False):

        if ts == True:
            ops = [qml.RY]  # time-symmetric unitaries
        else:
            ops = [qml.RX, qml.RY, qml.RZ]  # general unitaries

        weights = np.random.rand(layers, gates) * np.pi
        RandomLayers(weights, wires=range(qubits), rotations=ops, seed=np.random.randint(0, 10000))

        return [qml.sample(op=qml.PauliY(q)) for q in range(qubits)]

    return circuit
```

## Teste

```python
# the measurement outcomes for the first 3 shots
circuit = generate_circuit(n_shots)
print(np.array(circuit(ts=True))[:, 0:3])
print("\n")
print(np.array(circuit(ts=False))[:, 0:3])
```

## Gerando os operadores T-Simétricos

```python
def generate_circuits(circuits, shots):
    raw_data = []
    
    for ts in [True, False]:
        for __ in range(circuits):
            circuit = generate_circuit(n_shots)
            raw_data.append(circuit(ts=ts))
    return raw_data

raw_data = generate_circuits(circuits, n_shots)
```

## Visualização

```python
print(raw_data[0])
```

## Pré-processamento 

```python
def process_data(raw_data):
    "convert raw data to vectors of means and variances of each qubit"

    raw_data = np.array(raw_data)
    nc = len(raw_data)  # the number of circuits used to generate the data
    nq = len(raw_data[0])  # the number of qubits in each circuit
    new_data = np.zeros([nc, 2 * nq])

    for k, outcomes in enumerate(raw_data):
        means = [np.mean(outcomes[q, :]) for q in range(nq)]
        variances = [np.var(outcomes[q, :]) for q in range(nq)]
        new_data[k] = np.array(means + variances)

    return new_data


data = process_data(raw_data)
```

## Visualização

```python
print(data[0])
```

## Preparando e rodando o classificador

```python
kernel_pca = KernelPCA(
        n_components=None, kernel="rbf", gamma=None, fit_inverse_transform=True, alpha=0.1
    )

def runClass(data):
    # rescale the data so it has unit standard deviation and zero mean.
    scaler = preprocessing.StandardScaler().fit(data)
    data = scaler.transform(data)
    # try to cluster the data
    fit = kernel_pca.fit(data).transform(data)

    return fit

fit = runClass(data)
```

```python
# make a colour map for the points
c = np.array([0 for __ in range(circuits)] + [1 for __ in range(circuits)])

plt.scatter(fit[:, 0], fit[:, 1], c=c)
plt.show()
```

## Testando com o K-means

```python
def runCluster(data, n_clusters=2):
    # Normaliza os dados para média 0 e variância 1
    scaler = preprocessing.StandardScaler().fit(data)
    data_scaled = scaler.transform(data)  # Armazena os dados normalizados

    # Aplica K-Means
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(data_scaled)  # Obtém os rótulos dos clusters

    return data_scaled, labels  # Retorna os dados normalizados e os clusters

# Número de clusters esperado
n_clusters = 2  # Ajuste conforme necessário

data_scaled, clusters = runCluster(data, n_clusters)

pca = PCA(n_components=2)
data_2D = pca.fit_transform(data_scaled)

# Plota os clusters com cores baseadas nos rótulos do K-Means
plt.scatter(data_2D[:, 0], data_2D[:, 1], c=clusters, cmap="viridis")
plt.title("Clusters encontrados pelo K-Means")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.colorbar(label="Cluster")
plt.show()
```

## Função para gerar circuitos com memória quântica

```python
def CNOT_sequence(control_wires, target_wires):
    """Apply CNOTs in sequence using the provided control and target wires"""
    for c_wire, t_wire in zip(control_wires, target_wires):
        qml.CNOT([c_wire, t_wire])

def generate_ecircuit(n_shots):

    dev = qml.device("lightning.qubit", wires=qubits * 2, shots=n_shots)
    
    @qml.qnode(dev)
    def enhanced_circuit(ts=False):
        "implement the enhanced circuit, using a random unitary"
    
        if ts == True:
            ops = [qml.RY]
        else:
            ops = [qml.RX, qml.RY, qml.RZ]
    
        weights = np.random.rand(layers, n_shots) * np.pi
        seed = np.random.randint(0, 10000)
    
        for q in range(qubits):
            qml.Hadamard(wires=q)
    
        CNOT_sequence(control_wires=range(qubits), target_wires=range(qubits, 2 * qubits))
        RandomLayers(weights, wires=range(0, qubits), rotations=ops, seed=seed)
        RandomLayers(weights, wires=range(qubits, 2 * qubits), rotations=ops, seed=seed)
        CNOT_sequence(control_wires=range(qubits), target_wires=range(qubits, 2 * qubits))
    
        for q in range(qubits):
            qml.Hadamard(wires=q)
    
        return [qml.sample(op=qml.PauliZ(q)) for q in range(2 * qubits)]
        
    return enhanced_circuit
        
```

## Gerando circuitos

```python
def generate_ecircuits(circuits, n_shots):
    raw_data = []
    
    for ts in [True, False]:
        for __ in range(circuits):
            enhanced_circuit = generate_ecircuit(n_shots)
            raw_data.append(enhanced_circuit(ts))

    return raw_data

raw_data = generate_ecircuits(circuits, n_shots)
```

# Adicionando ruído

```python
def noise_layer(epsilon):
    "apply a random rotation to each qubit"
    for q in range(2 * qubits):
        angles = (2 * np.random.rand(3) - 1) * epsilon
        qml.Rot(angles[0], angles[1], angles[2], wires=q)

def generate_ecircuit(n_shots):

    dev = qml.device("lightning.qubit", wires=qubits * 2, shots=n_shots)
    
    @qml.qnode(dev)
    def enhanced_circuit(ts=False):
        "implement the enhanced circuit, using a random unitary"
    
        if ts == True:
            ops = [qml.RY]
        else:
            ops = [qml.RX, qml.RY, qml.RZ]
    
        weights = np.random.rand(layers, n_shots) * np.pi
        seed = np.random.randint(0, 10000)
    
        for q in range(qubits):
            qml.Hadamard(wires=q)
    
        CNOT_sequence(control_wires=range(qubits), target_wires=range(qubits, 2 * qubits))
        RandomLayers(weights, wires=range(0, qubits), rotations=ops, seed=seed)
        RandomLayers(weights, wires=range(qubits, 2 * qubits), rotations=ops, seed=seed)
        noise_layer(np.pi / 4)  # added noise layer
        CNOT_sequence(control_wires=range(qubits), target_wires=range(qubits, 2 * qubits))
    
        for q in range(qubits):
            qml.Hadamard(wires=q)
    
        return [qml.sample(op=qml.PauliZ(q)) for q in range(2 * qubits)]
        
    return enhanced_circuit

raw_data = []

for ts in [True, False]:
    for __ in range(circuits):
        enhanced_circuit = generate_ecircuit(n_shots)
        raw_data.append(enhanced_circuit(ts))

data = process_data(raw_data)

kernel_pca = KernelPCA(
    n_components=None, kernel="rbf", gamma=None, fit_inverse_transform=True, alpha=0.1
)
scaler = preprocessing.StandardScaler().fit(data)
data = scaler.transform(data)
fit = kernel_pca.fit(data).transform(data)

c = np.array([0 for __ in range(circuits)] + [1 for __ in range(circuits)])
plt.scatter(fit[:, 0], fit[:, 1], c=c)
plt.show()
```

## Calculo de distância quântica convencinal

```python
#Definição dos números complexos
A = 1 + 2j
B = 2 + 5j

# Convertendo para número complexo
theta_A = np.angle(A)  # Ângulo de fase de A
theta_B = np.angle(B)  # Ângulo de fase de B

# Fator de escala Z baseado nas normas dos vetores
Z = (abs(A) + abs(B)) / 2  

# Dispositivo quântico com 3 qubits
n_shots = 10000
dev = qml.device("lightning.qubit", wires=3, shots=n_shots)

@qml.qnode(dev)
def quantum_distance():
    # Passo 1: Aplicar Hadamard nos três primeiros qubits
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.Hadamard(wires=2)

    # Passo 2: Aplicar RY para codificar os ângulos (equivalente a U3 do Qiskit)
    qml.RY(2 * theta_A, wires=1)  # Multiplicamos por 2 para capturar fase corretamente
    qml.RY(2 * theta_B, wires=2)

    # Passo 3: Executar swap test
    qml.CSWAP(wires=[0, 1, 2])
    qml.Hadamard(wires=0)

    # Medição do primeiro qubit
    return qml.sample(qml.PauliZ(0))

# Executa o circuito
samples = quantum_distance()

# Contagem dos resultados
counts = {"+1": np.sum(samples == 1), "-1": np.sum(samples == -1)}
prob_0 = counts["-1"] / n_shots  # Probabilidade de medir 1 (estado '001')

# Distância Quântica baseada no swap test
D_Q = np.sqrt(1 - prob_0)  # Distância original do swap test
D_Q_scaled = Z * D_Q  # Distância escalada para aproximar a Euclidiana

# Distância Euclidiana real entre A e B
D_E = abs(A - B)

# Resultados
print(f"Probabilidade de medir '001': {prob_0:.4f}")
print(f"Quantum Distance (original): {D_Q:.4f}")
print(f"Quantum Distance (scaled): {D_Q_scaled:.4f}")
print(f"Euclidean Distance: {D_E:.4f}")
```

## Calculo de distância quântica melhorado

```python
# Definição dos números complexos
A = 1 + 2j
B = 2 + 5j

# Convertendo para número complexo
theta_A = np.angle(A)  # Ângulo de fase de A
theta_B = np.angle(B)  # Ângulo de fase de B

# Fator de escala Z baseado nas normas dos vetores
Z = (abs(A) + abs(B)) / 2  

# Dispositivo quântico com 6 qubits
n_shots = 10000
dev = qml.device("lightning.qubit", wires=6, shots=n_shots)

@qml.qnode(dev)
def quantum_distance_entangled():
    # Passo 1: Aplicar Hadamard nos três primeiros qubits
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.Hadamard(wires=2)

    # Passo 3: Criar pares emaranhados usando CNOTs
    qml.CNOT(wires=[0, 3])  # Segundo qubit controla o quarto
    qml.CNOT(wires=[1, 4])  # Terceiro qubit controla o quinto
    qml.CNOT(wires=[2, 5])  # Primeiro qubit controla o sexto

    # Passo 2: Aplicar RY para codificar os ângulos
    qml.RY(2 * theta_A, wires=1)  # Multiplicamos por 2 para capturar fase corretamente
    qml.RY(2 * theta_A, wires=4)
    qml.RY(2 * theta_B, wires=2)
    qml.RY(2 * theta_B, wires=5)

    # Passo 4: Executar swap test
    qml.CSWAP(wires=[0, 1, 2])

    # Passo 3: Criar pares emaranhados usando CNOTs
    qml.CNOT(wires=[0, 3])  # Segundo qubit controla o quarto
    qml.CNOT(wires=[1, 4])  # Terceiro qubit controla o quinto
    qml.CNOT(wires=[2, 5])  # Primeiro qubit controla o sexto

    qml.Hadamard(wires=0)

    # Medição do primeiro qubit
    return qml.sample(qml.PauliZ(0))

# Executa o circuito
samples_entangled = quantum_distance_entangled()

# Contagem dos resultados
counts_entangled = {"+1": np.sum(samples_entangled == 1), "-1": np.sum(samples_entangled == -1)}
prob_0_entangled = counts_entangled["-1"] / n_shots  # Probabilidade de medir 1 (estado '001')

# Distância Quântica baseada no swap test com emaranhamento
D_Q_entangled = np.sqrt(1 - prob_0_entangled)
D_Q_scaled_entangled = Z * D_Q_entangled  # Distância escalada para aproximar a Euclidiana

# Distância Euclidiana real entre A e B
D_E = abs(A - B)

# Resultados
print(f"Probabilidade de medir '001' (entangled): {prob_0_entangled:.4f}")
print(f"Quantum Distance (original, entangled): {D_Q_entangled:.4f}")
print(f"Quantum Distance (scaled, entangled): {D_Q_scaled_entangled:.4f}")
print(f"Euclidean Distance: {D_E:.4f}")
```

## Função para calcular a distância quânticamente

```python
# Definição das funções para calcular as distâncias
def quantum_distance(theta_A, theta_B):
    dev_no_ent = qml.device("lightning.qubit", wires=3, shots=n_shots)
    
    @qml.qnode(dev_no_ent)
    def circuit():
        qml.Hadamard(wires=0)
        qml.Hadamard(wires=1)
        qml.Hadamard(wires=2)

        qml.RY(2 * theta_A, wires=1)
        qml.RY(2 * theta_B, wires=2)

        qml.CSWAP(wires=[0, 1, 2])
        qml.Hadamard(wires=0)

        return qml.sample(qml.PauliZ(0))
    
    samples = circuit()
    prob_0 = np.sum(samples == -1) / n_shots
    return (abs(A) + abs(B)) / 2  * np.sqrt(1 - prob_0)
```

## Função para calcular a distância quânticamente melhorado

```python
def quantum_distance_entangled(theta_A, theta_B):
    dev = qml.device("lightning.qubit", wires=6, shots=n_shots)
    
    @qml.qnode(dev)
    def circuit():
        # Passo 1: Aplicar Hadamard nos três primeiros qubits
        qml.Hadamard(wires=0)
        qml.Hadamard(wires=1)
        qml.Hadamard(wires=2)
    
        # Passo 3: Criar pares emaranhados usando CNOTs
        qml.CNOT(wires=[0, 3])  # Segundo qubit controla o quarto
        qml.CNOT(wires=[1, 4])  # Terceiro qubit controla o quinto
        qml.CNOT(wires=[2, 5])  # Primeiro qubit controla o sexto
    
        # Passo 2: Aplicar RY para codificar os ângulos
        qml.RY(2 * theta_A, wires=1)  # Multiplicamos por 2 para capturar fase corretamente
        qml.RY(2 * theta_A, wires=4)
        qml.RY(2 * theta_B, wires=2)
        qml.RY(2 * theta_B, wires=5)
    
        # Passo 4: Executar swap test
        qml.CSWAP(wires=[0, 1, 2])
    
        # Passo 3: Criar pares emaranhados usando CNOTs
        qml.CNOT(wires=[0, 3])  # Segundo qubit controla o quarto
        qml.CNOT(wires=[1, 4])  # Terceiro qubit controla o quinto
        qml.CNOT(wires=[2, 5])  # Primeiro qubit controla o sexto
    
        qml.Hadamard(wires=0)
    
        # Medição do primeiro qubit
        return qml.sample(qml.PauliZ(0))
    
    samples = circuit()
    prob_0 = np.sum(samples == -1) / n_shots
    return (abs(A) + abs(B)) / 2  * np.sqrt(1 - prob_0)
```

## Avaliação de melhoria

```python
# Função para gerar pontos aleatórios e avaliar o erro
def evaluate_distance_model(n_samples=1000):
    errors_no_ent = []
    errors_ent = []
    errors_pct_no_ent = []
    errors_pct_ent = []
    
    for _ in range(n_samples):
        # Gera números complexos aleatórios A e B
        A = np.random.uniform(-5, 5) + 1j * np.random.uniform(-5, 5)
        B = np.random.uniform(-5, 5) + 1j * np.random.uniform(-5, 5)

        # Calcula distâncias
        D_E = abs(A - B)
        D_Q = quantum_distance(np.angle(A), np.angle(B))
        D_Q_ent = quantum_distance_entangled(np.angle(A), np.angle(B))

        # Calcula erros
        errors_no_ent.append((D_Q - D_E) ** 2)
        errors_ent.append((D_Q_ent - D_E) ** 2)

        errors_pct_no_ent.append(abs(D_Q - D_E) / D_E * 100)
        errors_pct_ent.append(abs(D_Q_ent - D_E) / D_E * 100)

    # Calcula métricas estatísticas
    mse_no_ent = np.sqrt(np.mean(errors_no_ent))
    mse_ent = np.sqrt(np.mean(errors_ent))

    mean_err_pct_no_ent = np.mean(errors_pct_no_ent)
    mean_err_pct_ent = np.mean(errors_pct_ent)

    print(f"Erro Médio Quadrático (Sem Emaranhamento): {mse_no_ent:.4f}")
    print(f"Erro Médio Quadrático (Com Emaranhamento): {mse_ent:.4f}")
    print(f"Erro Percentual Médio (Sem Emaranhamento): {mean_err_pct_no_ent:.2f}%")
    print(f"Erro Percentual Médio (Com Emaranhamento): {mean_err_pct_ent:.2f}%")

    # Plotando os erros
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.hist(errors_pct_no_ent, bins=20, alpha=0.7, label="Sem Emaranhamento")
    plt.hist(errors_pct_ent, bins=20, alpha=0.7, label="Com Emaranhamento")
    plt.xlabel("Erro Percentual (%)")
    plt.ylabel("Frequência")
    plt.legend()
    plt.title("Distribuição do Erro Percentual")

    plt.subplot(1, 2, 2)
    plt.boxplot([errors_pct_no_ent, errors_pct_ent], tick_labels=["Sem Emaranhamento", "Com Emaranhamento"])
    plt.ylabel("Erro Percentual (%)")
    plt.title("Boxplot do Erro Percentual")

    plt.tight_layout()
    plt.show()

# Executa a avaliação
evaluate_distance_model(n_samples=500)
```

# Aplicando em outros circuitos com ruido

## Função de calculo de distância quântica convencinal

```python
# Definição das funções para calcular as distâncias
def quantum_distance(theta_A, theta_B):
    dev_no_ent = qml.device("default.mixed", wires=3, shots=n_shots)
    
    @qml.qnode(dev_no_ent)
    def circuit():
        qml.Hadamard(wires=0)
        qml.Hadamard(wires=1)
        qml.Hadamard(wires=2)

        qml.RY(2 * theta_A, wires=1)
        qml.RY(2 * theta_B, wires=2)

        qml.CSWAP(wires=[0, 1, 2])
        
        # Adicionando ruído em pontos estratégicos
        qml.DepolarizingChannel(0.05, wires=1)  # Ruído despolarizante no qubit 1
        qml.DepolarizingChannel(0.05, wires=2)  # Ruído despolarizante no qubit 2
    
        qml.PhaseDamping(0.1, wires=0)  # Ruído de fase no qubit 0
        
        qml.Hadamard(wires=0)

        return qml.sample(qml.PauliZ(0))
    
    samples = circuit()
    prob_0 = np.sum(samples == -1) / n_shots
    return (abs(A) + abs(B)) / 2  * np.sqrt(1 - prob_0)
```

## Função de calculo de distância quântica melhorado

```python
def quantum_distance_entangled(theta_A, theta_B):
    dev = qml.device("default.mixed", wires=6, shots=n_shots)
    
    @qml.qnode(dev)
    def circuit():
        # Passo 1: Aplicar Hadamard nos três primeiros qubits
        qml.Hadamard(wires=0)
        qml.Hadamard(wires=1)
        qml.Hadamard(wires=2)
    
        # Passo 3: Criar pares emaranhados usando CNOTs
        qml.CNOT(wires=[0, 3])  # Segundo qubit controla o quarto
        qml.CNOT(wires=[1, 4])  # Terceiro qubit controla o quinto
        qml.CNOT(wires=[2, 5])  # Primeiro qubit controla o sexto
    
        # Passo 2: Aplicar RY para codificar os ângulos
        qml.RY(2 * theta_A, wires=1)  # Multiplicamos por 2 para capturar fase corretamente
        qml.RY(2 * theta_A, wires=4)
        qml.RY(2 * theta_B, wires=2)
        qml.RY(2 * theta_B, wires=5)
    
        # Passo 4: Executar swap test
        qml.CSWAP(wires=[0, 1, 2])

        # Adicionando ruído em pontos estratégicos
        qml.DepolarizingChannel(0.05, wires=1)  # Ruído despolarizante no qubit 1
        qml.DepolarizingChannel(0.05, wires=2)  # Ruído despolarizante no qubit 2
        qml.DepolarizingChannel(0.05, wires=4)
        qml.DepolarizingChannel(0.05, wires=5)
    
        qml.PhaseDamping(0.1, wires=0)  # Ruído de fase no qubit 0
        qml.PhaseDamping(0.1, wires=3)
    
        # Passo 3: Criar pares emaranhados usando CNOTs
        qml.CNOT(wires=[0, 3])  # Segundo qubit controla o quarto
        qml.CNOT(wires=[1, 4])  # Terceiro qubit controla o quinto
        qml.CNOT(wires=[2, 5])  # Primeiro qubit controla o sexto
    
        qml.Hadamard(wires=0)
    
        # Medição do primeiro qubit
        return qml.sample(qml.PauliZ(0))
    
    samples = circuit()
    prob_0 = np.sum(samples == -1) / n_shots
    return (abs(A) + abs(B)) / 2  * np.sqrt(1 - prob_0)
```

## Avaliação de melhoria

```python
# Função para gerar pontos aleatórios e avaliar o erro
def evaluate_distance_model(n_samples=1000):
    errors_no_ent = []
    errors_ent = []
    errors_pct_no_ent = []
    errors_pct_ent = []
    
    for _ in range(n_samples):
        # Gera números complexos aleatórios A e B
        A = np.random.uniform(-5, 5) + 1j * np.random.uniform(-5, 5)
        B = np.random.uniform(-5, 5) + 1j * np.random.uniform(-5, 5)

        # Calcula distâncias
        D_E = abs(A - B)
        D_Q = quantum_distance(np.angle(A), np.angle(B))
        D_Q_ent = quantum_distance_entangled(np.angle(A), np.angle(B))

        # Calcula erros
        errors_no_ent.append((D_Q - D_E) ** 2)
        errors_ent.append((D_Q_ent - D_E) ** 2)

        errors_pct_no_ent.append(abs(D_Q - D_E) / D_E * 100)
        errors_pct_ent.append(abs(D_Q_ent - D_E) / D_E * 100)

    # Calcula métricas estatísticas
    mse_no_ent = np.mean(errors_no_ent)
    mse_ent = np.mean(errors_ent)

    mean_err_pct_no_ent = np.mean(errors_pct_no_ent)
    mean_err_pct_ent = np.mean(errors_pct_ent)

    print(f"Erro Médio Quadrático (Sem Emaranhamento): {mse_no_ent:.4f}")
    print(f"Erro Médio Quadrático (Com Emaranhamento): {mse_ent:.4f}")
    print(f"Erro Percentual Médio (Sem Emaranhamento): {mean_err_pct_no_ent:.2f}%")
    print(f"Erro Percentual Médio (Com Emaranhamento): {mean_err_pct_ent:.2f}%")

    # Plotando os erros
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.hist(errors_pct_no_ent, bins=20, alpha=0.7, label="Sem Emaranhamento")
    plt.hist(errors_pct_ent, bins=20, alpha=0.7, label="Com Emaranhamento")
    plt.xlabel("Erro Percentual (%)")
    plt.xlim(-10, 1000)
    plt.ylabel("Frequência")
    plt.legend()
    plt.title("Distribuição do Erro Percentual")

    plt.subplot(1, 2, 2)
    plt.boxplot([errors_pct_no_ent, errors_pct_ent], tick_labels=["Sem Emaranhamento", "Com Emaranhamento"])
    plt.ylabel("Erro Percentual (%)")
    plt.ylim(-10, 800)
    plt.title("Boxplot do Erro Percentual")

    plt.tight_layout()
    plt.show()

# Executa a avaliação
evaluate_distance_model(n_samples=500)
```
